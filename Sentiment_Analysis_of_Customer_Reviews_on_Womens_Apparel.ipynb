```python
# prompt: Read data from /content/sample_data/Womens Clothing E-Commerce Reviews.csv and show the head

import pandas as pd

df = pd.read_csv('/content/sample_data/Womens Clothing E-Commerce Reviews.csv')

print(df.head())
```

```python
# prompt: Create a word cloud based on "Review Text" after removing the stop words and lemmatizing the words.

!pip install wordcloud matplotlib nltk

import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Remove stop words and lemmatize
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    if isinstance(text, str):
        words = text.lower().split()
        words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
        return ' '.join(words)
    return ''

df['Processed Review Text'] = df['Review Text'].apply(preprocess_text)

# Create the word cloud
text = " ".join(review for review in df['Processed Review Text'])
wordcloud = WordCloud(stopwords=stop_words, background_color="white").generate(text)

# Display the word cloud
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# prompt: Do a sentiment analysis on the "Review Text."

!pip install nltk textblob
import nltk
from textblob import TextBlob
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

def get_sentiment(text):
    if isinstance(text, str):
        analysis = TextBlob(text)
        # Return the polarity score
        return analysis.sentiment.polarity
    return 0.0

df['Sentiment'] = df['Review Text'].apply(get_sentiment)

print(df[['Review Text', 'Sentiment']].head())

# You can further analyze the sentiment column, e.g., categorize into positive, negative, neutral
def categorize_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment Category'] = df['Sentiment'].apply(categorize_sentiment)

print(df['Sentiment Category'].value_counts())
```

```python
# prompt: Create a Pie Chart for the sentiment category

import matplotlib.pyplot as plt
# Plot a pie chart of sentiment categories
sentiment_counts = df['Sentiment Category'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Sentiment Categories')
plt.show()
```

```python
# prompt: Classify the "Review Text" by Department and show the sentiment analysis in a bar chart.

import matplotlib.pyplot as plt
# Analyze sentiment by Department Name
sentiment_by_department = df.groupby('Department Name')['Sentiment'].mean().sort_values()

# Plot sentiment by department
plt.figure(figsize=(12, 6))
sentiment_by_department.plot(kind='bar', color='skyblue')
plt.title('Average Sentiment by Department')
plt.xlabel('Department Name')
plt.ylabel('Average Sentiment Score (TextBlob Polarity)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: Classify the "Review Text" by Department and show the postive, negative and neutral sentiment analysis in a clustered bar chart.

import matplotlib.pyplot as plt
import seaborn as sns

# Group by Department Name and Sentiment Category
sentiment_department_counts = df.groupby(['Department Name', 'Sentiment Category']).size().unstack(fill_value=0)

# Plot the clustered bar chart
plt.figure(figsize=(14, 8))
sentiment_department_counts.plot(kind='bar', stacked=False, figsize=(14, 8))
plt.title('Sentiment Distribution by Department')
plt.xlabel('Department Name')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sentiment Category')
plt.tight_layout()
plt.show()
```
